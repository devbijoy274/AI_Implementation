# Programm to get into Linear Regression
Linear regression with one variable to predict profits for a food truck. Suppose you are the CEO of a restaurant franchise and are considering different cities for opening a new outlet. The chain already has trucks in various cities and you have data for profits and populations from the cities.
You would like to use given data to help you select which city to expand
to next.

```
import tensorflow as tf
import pandas as pd

data = pd.read_csv('ex1data1.txt', sep=',', header=None)
data.columns = ['X','y']
print(data)
```

```
#Scatterplot data visualization
import matplotlib.pyplot as plt

X = data.X
y = data.y
plt.plot(X,y,'bo',ms=6)
plt.ylabel('Profit in $10,000s')
plt.xlabel('Population of City in 10,000s')
```

```
#Implementation to fit theta in LR
import numpy as np

alpha = 0.01
itr = 1500
m = len(X)
theta = np.zeros((2,1))
ones_X = np.ones((m,1))
X = np.array(X)
X = X.reshape(m,1)
X = np.column_stack((ones_X,X))
X = np.asmatrix(X)
y = y.reshape(m,1)
y = np.asmatrix(y)

for i in range(itr):
    x0 = X[:,0]
    x1 = X[:,1]
    hxX =  ((X * theta) - y)
    temp1 = (theta[0] - ((alpha/m) * (np.sum(np.multiply(hxX,x0)))))                  
    temp2 = (theta[1] - ((alpha/m) * (np.sum(np.multiply(hxX,x1)))))
    theta[0] = temp1
    theta[1] = temp2
    hx = X * theta
    err = hx - y
    jTheta = (1/(2*m) * (np.sum(np.square(err))))
    print(jTheta)
```
